```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(x):
    vif=pd.DataFrame()
    vif["variables"]=x.columns
    vif["VIF"]=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
    return(vif)
    
```

```python
#importing VIF from statsmodel
#creating calc_vif function with the input of X
#initializing an empty DataFrame of vif to store the values
#creating two columns of vif 1.variables 2.VIF
#assign the input(X)columns to the variables columns
#for loop executing,in that Input columns (shape[1]=columns,shape[0]=Rows)looping one by one to i
#VIF calculating for the x.values and i by variance_inflation_factor 
#calculated vif values stored in the list.
#function returns the vif values to the DataFrame
```

```python
#HYPOTHESIS TESTING

#independent sample-unpaired TTest
#different group(male,female) but same condition(salary)
from scipy.stats import ttest_ind
dataset=dataset.dropna()
male=dataset[dataset["gender"]=="M"]["salary"]
female=dataset[dataset["gender"]=="F"]["salary"]
ttest_ind(male,female)
```

```python
#independent sample-unpaired TTest
#different group(others,central) but same condition(salary)
from scipy.stats import ttest_ind
#dataset=dataset.dropna()
male=dataset[dataset["hsc_b"]=="Others"]["salary"]
female=dataset[dataset["hsc_b"]=="Central"]["salary"]
ttest_ind(male,female)
```

```python
#dependent sample-paired TTest
#same group(male) but different condition(ssc_p,hsc_p)
from scipy.stats import ttest_rel
#dataset=dataset.dropna()
male=dataset[dataset["gender"]=="M"]["ssc_p"]
male1=dataset[dataset["gender"]=="M"]["hsc_p"]
ttest_rel(male,male1)
```

```python
#dependent sample-paired TTest
#same group(commerce) but different condition(etest_p,mba_p)
from scipy.stats import ttest_ind
dataset=dataset.dropna()
#dataset=dataset.dropna()
etest=dataset[dataset["hsc_s"]=="commerce"]["etest_p"]
mba=dataset[dataset["hsc_s"]=="commerce"]["mba_p"]
ttest_ind(etest,mba)
```

```python
#ANAVO :Analysis of variance
#one way classification
import scipy.stats as stats
stats.f_oneway(dataset["ssc_p"],dataset["hsc_p"],dataset["degree_p"])
```

```python
#statement:there is significant difference between ssc_p,hsc_p,degree_p
#Ho:There is no significant difference 
#H1:There is significant difference
#Reject Null hypothesis  if p<0.05
#Accept Null hypothesis and reject alternate hypothesis
#RESULT
# Here pvalue=0.51(0.51<0.05)
#H1:There is significant difference between ssc_p,hsc_p,degree_p
# So we accept null hypothesis and reject alternate
```
