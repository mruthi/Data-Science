```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
dataset["salary"].fillna(0,inplace=True)
```

```python
dataset
```

```python
import matplotlib.pyplot as plt
x=dataset["mba_p"]
y=dataset["salary"]
plt.plot(x,y)
plt.title("LINE PLOT",COLOR="blue")
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=dataset["mba_p"]
y=dataset["salary"]
plt.scatter(x,y)
plt.title("SCATTER PLOT",COLOR="blue")
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=dataset["specialisation"]
y=dataset["salary"]
plt.bar(x,y)
plt.title("BAR PLOT",COLOR="green")
plt.xlabel('X-axis',color="red")
plt.ylabel('y-axis',color="purple")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=dataset["mba_p"]
y=dataset["salary"]
plt.fill_between(x,y)
plt.title("AREA PLOT",COLOR="blue")
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
x=np.random.standard_normal(100)
y=np.random.standard_normal(100)
z=np.random.standard_normal(100)
ax.scatter(x,y,z)
ax.set_xlabel('X-axis')
ax.set_xlabel('Y-axis')
ax.set_xlabel('Z-axis')
plt.title('3D Scatter Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randn(1000)
y=np.random.randn(1000)
plt.hexbin(x,y,gridsize=30,cmap='viridis')
plt.colorbar()
plt.title("Hexin plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
dataset["gender"].value_counts().plot(kind="pie",autopct='%1.1f%%')
plt.title("pie chart",color="green")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data=[np.random.normal(0,std,100)for std in range(1,4)]
plt.violinplot(data)
plt.title("VIOLIN PLOT",COLOR="blue")
plt.show()
```

```python
import matplotlib.pyplot as plt
dataset["salary"].plot(kind="hist",bins=10,title="HISTOGRAM",COLOR="blue")
```

```python
import matplotlib.pyplot as plt

dataset.boxplot(column="ssc_p",by='status',figsize=(8,6))
title("BOX PLOT",COLOR="blue")
plt.show()
```
