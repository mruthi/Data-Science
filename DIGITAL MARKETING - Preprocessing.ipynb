```python
import pandas as pd
```

```python
data=pd.read_csv("Inputdata.csv")
```

```python
data.isnull().sum()
```

```python
data.fillna(0,inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
new=data.groupby('Campaign name').apply(lambda x:x.sort_values(by="Day",ascending=True))
```

```python
new
```

```python
data.columns
```

```python
output=pd.DataFrame(columns=["Campaign name","Date","AD Spent","No of Leads","Cost Per Lead","CTR","Link Click","LP Views","Member Passed","LP Conversion %"])
for columnName in new:
    output["Campaign name"]=new["Campaign name"]
    output["Date"]=new["Day"]
    output["AD Spent"]=new["Amount spent (INR)"]
    output["No of Leads"]=new["Results"]
    output["Cost Per Lead"]=new["Amount spent (INR)"]/new["Results"]
    output["CTR"]=new["CTR (link click-through rate)"]
    output["Link Click"]=new["Link clicks"]
    output["LP Views"]=new["Landing page views"]
    output["Member Passed"]=(new["Landing page views"]/new["Link clicks"])*100
    output["LP Conversion %"]=(new["Results"]/new["Landing page views"])*100
    
```

```python
output
```

```python
output.fillna(0,inplace=True)
```
