# QUESTION 8:Test the similariy between the degree_t(Sci&Tech)and specialisation(Mkt&HR)with respect
# to salary significance level of 5%.(Decision using Hypothesis Testing)

```python
#independent sample-unpaired TTest
#different group(degree_t(sci&Tech),specialisation(Mkt&HR)) but same condition(salary)
from scipy.stats import ttest_ind
dataset=dataset.dropna()
D=dataset[dataset["degree_t"]=="Sci&Tech"]["salary"]
S=dataset[dataset["specialisation"]=="Mkt&HR"]["salary"]
ttest_ind(D,S)
```


# Ho:There is no significant similarity(Null hypothesis)
# H1:There is significant similarity (Alternate hypothesis)
# p value=0.007 which is <0.05,that we accept alternate hypothesis(H1).So,There is significant similarity between degree_t(sci&Tech) and specialisation(Mkt&HR) with salary

# QUESTION 9:Convert the normal distribution to standard normal distribution for salary column

```python
def stdNBgraph(dataset):
    #converting to std normal distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    values=[i for i in dataset]
    z_score=[((j-mean)/std)for j in values]
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
```

```python
stdNBgraph(dataset["salary"])
```

# QUESTION 10:What is the Probability Density Function of the salary range from 7 lakhs to 9 lakhs?

```python
def get_pdf_probability(dataset,startrange,endrange):
    #importing needed libraries
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    #kde=kernal density fn.have to give kde=true.then only we can make curve.
    #mentioned below as curve line is blue,histogram color is green
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='green')
    # axis vertical line startrange color is red,endrange line color also red
    #startrange and endrange value getting from the user
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    #generate a sample
    sample = dataset
    #calculating mean and median 
    sample_mean =sample.mean()
    sample_std =sample.std()
    #printing the values of mean and median
    print('Mean=%.3f, Standard Deviation=%.3f' %(sample_mean, sample_std))
    #define the distribution
    #having default fn for norm
    #its learnig how to distribute from getting mean ,median values by using norm
    dist = norm(sample_mean, sample_std)
    #sample probabilities for a range pf outcomes
    #(values = []
    #for value in range(startrange, endrange):
       # values.append(value))
    #for the above method used the below oneliner for loop method
    values = [value for value in range(startrange, endrange)]
    #values converted by using pdf fn and transformed to probabilities
    probabilities = [dist.pdf(value) for value in values]
    #creaing summation for probabilities
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset["salary"],700000,900000)
```

# RESULT:There is 0.0005% Probability to get the salary range of 7 lakhs to 9 lakhs.

# QUESTION 11:Test the similarity between the degree_t(Sci&Tech)with respect
# to etest_p and mba_p at significance level of 5%.(Decision using by Hypothesis Testing)

```python
#dependent sample-paired TTest
#same group(degree_t(sci&Tech)) but different condition(etest_p,mba_p)
from scipy.stats import ttest_rel
#dataset=dataset.dropna()
test=dataset[dataset["degree_t"]=="Sci&Tech"]["etest_p"]
test1=dataset[dataset["degree_t"]=="Sci&Tech"]["mba_p"]
ttest_rel(test,test1)
```


# Ho:There is no significant similarity (Null hypothesis)
# H1:There is significant similarity (Alternate hypothesis)
# p value=5.51 which is >0.05,that we accept null hypothesis(Ho).So,There is no significant similarity between degree_t(sci&Tech) and specialisation(Mkt&HR) with salary
