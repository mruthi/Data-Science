```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
dataset.dropna(inplace=True)
```

```python
dataset
```

```python
#RESULT:If we using this option ,we have to delete the entire Rows,so,there is a chance to loss the
#        data's in the dataset.It is a wrong method.
```

```python
#OPTION 2:If we using this for this dataset,it will consider as Non placed students are salaried.
#RESULT  :So,it is a wrong method.
```

```python
#OPTION 3:
```

```python
dataset["salary"].fillna(dataset["salary"].mean(),inplace=True)
```

```python
dataset
```

```python
#RESULT  :This method also  consider as Non placed students are salaried.So,it is also a wrong method.
```

```python
#OPTION 4:
```

```python
dataset["salary"].fillna(0,inplace=True)
```

```python
dataset
```

```python
#RESULT:In this method,we filled 0 irrespecive of NaN values for the Not placed students.
#       Without data loss,filled the missing values.So,this is a correct method of replacing the Nan values.
```
